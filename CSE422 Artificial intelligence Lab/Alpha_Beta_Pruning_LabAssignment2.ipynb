```python
import random
ID = input("Enter your student id: ")
range1 = input("Minimum and Maximum value for the range of negative HP: ").split()
range2 = list(map(int, range1))
x = list(ID)
id = list(map(int,x))
outputList = []
turns = id[0]
branches = id[2]
depth = 2*turns
leafNumber = branches**depth
sttr = ID[-2:][::-1]
hp = int(sttr)
outputList.append(f'1. Depth and Branches ratio {depth}:{branches}')

leaf = []
for i in range(0, leafNumber): #random leaf numbers within the given range
    n = random.randint(range2[0],range2[1])
    leaf.append(n)
numstring = ''
for i in leaf:
    numstring += ' ' +str(i)
outputList.append(f'2. Terminal States(Leaf Nodes) are{numstring}.')

graphList=[]
c = 0
for i in range(0, depth):
    g =[]
    for j in range(branches**i): #creating node (number of nodes = branches**depth) for each depth
        g.append(c)
        c+=1
    graphList += [g]
graphList += [leaf] 

dic = {}
for i in reversed(range(len(graphList))):
    parent = i-1
    if parent == -1:
        break
    div = len(graphList[i]) // branches #dividing leafs/nodes into their branches to connect with parent node
    for j in range(div): #j will be number of parents
        if div != 1:
            child = graphList[i][:branches]
            value = graphList[parent][j] #parent
            dic[value] = child #creating parent child dictionary
            del graphList[i][:branches]
        else:
            child = graphList[i]
            value = graphList[parent][j]
            dic[value] = child
            del graphList[i]

graph = {}
sortt = sorted(dic.keys()) #sorting the key number for convenience
for key in sortt:
    if key not in graph.keys():
        graph[key] = dic[key]

nodeVisited=[]
MAX, MIN = 1000, -1000
def AlphaBeta(Gdepth, node, maximizingPlayer, alpha, beta):
    if Gdepth == depth:
        return 'terminate'
    if maximizingPlayer: 
        best = MIN
        for i in range(len(graph[node])):
            val = AlphaBeta(Gdepth + 1, graph[node][i] , False, alpha, beta)
            if val == 'terminate':
                val = graph[node]
            if len(val) == 1: 
                val = val[-1]
            else:
                val = val[i]
            if Gdepth == depth-1:
                nodeVisited.append(val)
            best = max(best, val)
            alpha = max(alpha, best)
            if beta <= alpha:
                break
        return [best]
    else:
        best = MAX
        for i in range(len(graph[node])):
            val = AlphaBeta(Gdepth + 1, graph[node][i], True, alpha, beta)
            if val == 'terminate':
                val = graph[node]
            if len(val) == 1: 
                val = val[-1]
            else:
                val = val[i]
            if Gdepth == depth-1:
                nodeVisited.append(val)
            best = min(best, val)
            beta = min(beta, best)
            if beta <= alpha:
                break
        return [best]

damage = AlphaBeta(0, 0, True, MIN, MAX)[0]
comp = len(nodeVisited)
life = hp - damage
outputList.append(f'3. Left life(HP) of the defender after maximum damage caused by the attacker is {life}')
outputList.append(f'4. After Alpha-Beta Pruning Leaf Node Comparisons {comp}')

for i in outputList:
    print(i)
```
