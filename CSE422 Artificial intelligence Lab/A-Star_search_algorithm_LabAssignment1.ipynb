```python
import heapq
start = input('Start node: ').strip().capitalize()
end = input('Destination: ').strip().capitalize()
f = open('lab1input.txt', mode='r') #reading from notepad (txt file) named "lab1input"
lines=f.readlines()
HeurDict = {}
graph = {}
notations = {'Arad': 'A', 'Neamt' :'F', 'Bucharest': 'Z', 'Oradea' :'B', 'Craiova' : 'S', 'Pitesti' :'P', 'Eforie': 'T', 'RimnicuVilcea': 'R', 'Fagaras': 'O', 'Timisoara': 'C', 'Dobreta': 'V', 'Urziceni': 'D', 'Hirsova' : 'N','Vaslui' : 'H', 'lasi': 'Q', 'Zerind' : 'E', 'Lugoj': 'G', 'Mehadia': 'L', 'Giurgiu': 'X', 'Sibiu': 'Y'}

destinationSymbol = notations[end]
startSymbol = notations[start]

for i in lines:
    x = i.strip()
    list1 = x.split()
    node = list1[0] #graph first node
    valueH = int(list1[1])
    nodeSymbol = notations[node]
    HeurDict[nodeSymbol] = valueH #putting all heuristic values in a dictionary.
    list1.pop(0)
    list1.pop(0)

    loop = len(list1) // 2
    for j in range(loop):
        c1 = []
        cost = int(list1[1])
        vertex = list1[0]
        vertexSymbol = notations[vertex]
        c1 += [cost, vertexSymbol]  #storing the neighbor node and its pathcost as a list,
        if nodeSymbol in graph.keys(): # where node is the key and the [path cost, neighbor node] is the value in the dictionary.
            graph[nodeSymbol].append(c1)
        else:
            graph[nodeSymbol] = [c1]
        list1.pop(0)
        list1.pop(0)   

queue = []
goal = {} #all routes that have arrived at the destination and its total cost
allBranches = {} #all branches and its pathcost
def Astar(G, src):
    queue = [[0, src]]
    while queue:
        m = heapq.heappop(queue)
        branch = m[1] 
        branching = ''     #storing single node or a string of nodes  
        for cost, neighbour in graph.get(branch[-1]):       #only checking the latest node Symbol of a branch and its neighboring nodes.
            if neighbour not in branch:     #a branch must have distinct nodes.
                branching = branch
                branching += neighbour
                if branch in allBranches.keys():
                    allBranches[branching] = allBranches[branch] + cost   #updating pathcost 
                else:
                    allBranches[branching] = cost
                pathcost = allBranches[branching]
                Heuristic = HeurDict[neighbour]
                total = pathcost + Heuristic
                if neighbour == destinationSymbol:
                    goal[branching] = total
                else:
                    heapq.heappush(queue, [total, branching])
                    heapq.heapify(queue) 

Astar(graph, startSymbol)

outputList = []
if bool(goal) == True:
    result = min(goal.values()) #optimal value
    goalPath = []
    for k, v in goal.items():
        if v == result:
            goalPath.append(k)  #storing all potential routes with the optimal value

    notations1 = {}
    for k,v in notations.items():
        notations1[v] = k

    for i in goalPath:
        string = ''
        for j in i:
            if j in notations1.keys():
                string += notations1[j]
                string += '->'
        string = string[:-2] + '\n'
        string += f'Total distance: {result} km\n'
        outputList.append(string)
else:
    p1 = f'NO PATH FOUND'
    outputList.append(p1)

for i in outputList:
    print(i)
```
