```python
import random

inputFlag = False
length = int(input('The number of daily transactions: '))
if 1 > length or length > 100:
  print('Invalid input range')
  inputFlag = True

else:
  l1=[]
  for i in range(length):
      xy = input("Enter Amount: ").split()
      if 1 > int(xy[1]) or int(xy[1]) > 100000:
        print('Invalid input range')
        inputFlag = True
        break
      l1.append(xy)

if inputFlag!= True:
  for i in l1:
      if i[0].lower() == 'd':
          i[1] = int(i[1])
      if i[0].lower() == 'l':
          i[1] = -int(i[1])

  binary = []
  numofstrings = 6
  while True:
      bin = ''
      while True:
          n = str(random.randint(0,1))
          bin += n
          if len(bin) == length:
              break
      if bin.count('0') != length:
          binary.append(bin)
      if len(binary) == numofstrings:
          break

  result = 0
 
  for itr in range(0,5000):
      def fitness(value, string1):
          fd = {} #storing the fitness of all the strings in a dictionary 
          for string in string1:
              sum = 0
              for j in range(len(string)):
                  if string[j] == '1': #choosing index containing the string 1
                      amount = value[j][1]
                      sum += amount
              fd[string] = abs(sum)
          return fd

      fitnessdic = fitness(l1, binary)    

      invalid = '0'*length 
      for k,v in fitnessdic.items():
          if v == 0 and k != invalid:
              result = k

      if result != 0:
          break

      max_value = max(fitnessdic.values()) 
      max_key = 0
      for k,v in fitnessdic.items():
          if v == max_value:
              max_key = k
              break

      #removing the key with the greatest value of fitness/furthest away from the target (0)
      while True:
          binary.remove(max_key)
          if max_key not in binary:
              break

      #replacing with lowest value 
      if len(binary) != numofstrings:
          min_value = min(fitnessdic.values())
          min_key = 0
          for k,v in fitnessdic.items():
              if v == min_value:
                  min_key = k
                  break

      while True:
          binary.append(min_key)
          if len(binary) == numofstrings:
              break

      random.shuffle(binary)

      def crossover(string2):
          newC = []
          loop = len(string2)//2
          for i in range(loop):
              indexs = random.randint(0, length-1)
              b1, b2 = string2[0], string2[1]
              b1left, b1right = b1[:indexs], b1[indexs:]
              b2left, b2right = b2[:indexs], b2[indexs:]
              cross1 = b1left + b2right
              cross2 = b2left + b1right
              newC.append(cross1)
              newC.append(cross2)
              string2.pop(0)
              string2.pop(0)

          return newC

      crossstring = crossover(binary)

      def mutation(string3):
          newM = []
          for string in string3:
              s=[]
              for i in string:
                  s+=[i]
              index = random.randint(0, length-1)
              dig = str(random.randint(0, 1))
              s[index] = dig
              temp = ''
              for j in s:
                  temp+=j
              newM.append(temp)
            
          return newM

      mutationstring = mutation(crossstring)

      binary = mutationstring

  if result == 0:
      print('-1')
  else: 
      print(result)
```
